```python
import pandas as pd
from tqdm import tqdm_notebook

prefix = 'data/'
```

```python
df = pd.read_csv(prefix + 'sentiment_dataset.csv', header=None, encoding='latin-1')
df.head()
```

```python
train_df = df.sample(frac=0.8, random_state=200) #random state is a seed value
train_df.head()
```

```python
train_df1 = df.sample(frac=0.1, random_state=200) #random state is a seed value
train_df1.head()
```

```python
test_df = df.drop(train_df.index)
test_df.head()
```

```python
test_df1 = df.drop(train_df1.index).sample(frac=0.025, random_state=200)
test_df1.head()
```

```python
train_df[0] = train_df[0].apply(lambda x: 1 if(x == 4) else x)
train_df.head()
```

```python
train_df1[0] = train_df1[0].apply(lambda x: 1 if(x == 4) else x)
train_df1.head()
```

```python
test_df[0] = test_df[0].apply(lambda x: 1 if(x == 4) else x)
test_df.head()
```

```python
test_df1[0] = test_df1[0].apply(lambda x: 1 if(x == 4) else x)
test_df1.head()
```

```python
train_df = pd.DataFrame({
    'id':range(len(train_df)),
    'label':train_df[0],
    'alpha':['a']*train_df.shape[0],
    'text': train_df[5].replace(r'\n', ' ', regex=True)
})

train_df.head()
```

```python
train_df1 = pd.DataFrame({
    'id':range(len(train_df1)),
    'label':train_df1[0],
    'alpha':['a']*train_df1.shape[0],
    'text': train_df1[5].replace(r'\n', ' ', regex=True)
})

train_df1.head()
```

```python
dev_df = pd.DataFrame({
    'id':range(len(test_df)),
    'label':test_df[0],
    'alpha':['a']*test_df.shape[0],
    'text': test_df[5].replace(r'\n', ' ', regex=True)
})

dev_df.head()
```

```python
dev_df1 = pd.DataFrame({
    'id':range(len(test_df1)),
    'label':test_df1[0],
    'alpha':['a']*test_df1.shape[0],
    'text': test_df1[5].replace(r'\n', ' ', regex=True)
})

dev_df1.head()
```

```python
train_df.to_csv('data/train.tsv', sep='\t', index=False, header=False, columns=train_df.columns)
dev_df.to_csv('data/dev.tsv', sep='\t', index=False, header=False, columns=dev_df.columns)
```

```python
train_df1.to_csv('data/train1.tsv', sep='\t', index=False, header=False, columns=train_df1.columns)
dev_df1.to_csv('data/dev1.tsv', sep='\t', index=False, header=False, columns=dev_df1.columns)
```
